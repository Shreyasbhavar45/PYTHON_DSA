## __4.PMI AND RECURSION__

- Base case → starting proof
- Recursive call → assumption
- Multiply/add → k → k+1 transition

```python
##finding 2^n for a given __n__ using  recursion
def power2(n):

    #base condition
    if (n == 1):
        return 2
    

    ##assume:---> (Induction Hypothesis)
    ##Induction Hypothesis = “maan lena ki statement kisi ek value ke liye already true hai”
    smallAns = power2(n-1)
    Ans = 2 * smallAns
    return Ans


print(power2(5))

##flow of this answer
# power2(5)
# → 2 * power2(4)
# → 2 * (2 * power2(3))
# → 2 * (2 * (2 * power2(2)))
# → 2 * (2 * (2 * (2 * power2(1))))
# → 2 * 2 * 2 * 2 * 2
# → 32
```

## __Fibonnaci Series__

```python
##function Definition
def fibonnaciSeries(n):

    #base condition
    if n <= 1:
        return n
    
    else:
        return fibonnaciSeries(n-1)+ fibonnaciSeries(n-2)
##driver code
n = 4
for i in range(n):
    #function calling
    print(fibonnaciSeries(i))
```

## __Count Of number of Ways to reach Upstair__

```python
#helper function
def helper(n):
    if n<=1:
        return n
    return helper(n-1) + helper(n-2)

#function definition
def countNumberOfWays(n):
    return helper(n+1)
    
##drivercode
n = 5
result = countNumberOfWays(n)
print('Count Of number of Ways to reach Upstair:-',result)
```

```python
#2nd method
def countNumberOfWays(n):
    if n == 0 or n == 1:
        return 1
    
    return countNumberOfWays(n-1)+countNumberOfWays(n-2)

n = 5
result = countNumberOfWays(n)
print('Count Of number of Ways to reach Upstair:-',result)
```

## __Head vs Tail Recursion__

```python
##head recursion
def factHead(n):
    if (n==0):
        return 1
    
    smallanswer = factHead(n-1)
    finalanswer = n * smallanswer

    return finalanswer

print(factHead(5))
```

```python
##Tail Recursion
def factTail(n,accumulator = 1):
    if(n==0):
        return accumulator
    
    accumulator = n * accumulator
    
    return factTail(n-1,accumulator)


factTail(5)
```

## __Print Number 1 To N.__


```python
def print1ToN(n):
    if n <= 0:
        return []
    
    return print1ToN(n-1) + [n]


print1ToN(6)
```
