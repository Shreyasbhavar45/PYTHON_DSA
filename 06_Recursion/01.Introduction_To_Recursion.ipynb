## __4.PMI AND RECURSION__

- Base case → starting proof
- Recursive call → assumption
- Multiply/add → k → k+1 transition

```python
##finding 2^n for a given __n__ using  recursion
def power2(n):

    #base condition
    if (n == 1):
        return 2
    

    ##assume:---> (Induction Hypothesis)
    ##Induction Hypothesis = “maan lena ki statement kisi ek value ke liye already true hai”
    smallAns = power2(n-1)
    Ans = 2 * smallAns
    return Ans


print(power2(5))

##flow of this answer
# power2(5)
# → 2 * power2(4)
# → 2 * (2 * power2(3))
# → 2 * (2 * (2 * power2(2)))
# → 2 * (2 * (2 * (2 * power2(1))))
# → 2 * 2 * 2 * 2 * 2
# → 32
```
