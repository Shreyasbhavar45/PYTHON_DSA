## Introduction to Algorithms And Apriori Analysis .

# #What is algorithm?
- sequence of finite steps to perform same specific task.[**finite steps = it must end**]

# properties..
- 1) It should terminate after finite amount of time.
- 2) Produce atleast one output. - eg:-(1)Multiplication (2)Division.
- 3) Independent of any programming language.
- 4) Unambigious -> Deterministic:- Whenever you run the code it give answer.
