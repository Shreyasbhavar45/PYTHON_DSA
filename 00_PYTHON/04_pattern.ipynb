```python
#right angled Triangle
n = 6
#row 
for i in range(1,n+1):#outerloop
    #column
    for j in range(1,i+1):#innerloop
        print("*",end = " ")
    print()
```

```python
#Decreasing star pattern
n = 6
#row
for i in range(1,n+1):#outerloop
    #column
    for j in range(n,i-1,-1):
        print('*',end = ' ')
    print()
```

```python
#✅ Pattern 4 (Right-Aligned Triangle)
n = 9
for i in range(1,n+1):

    #spaces
    for s in range(n-i):
        print(' ',end=" ")

    #stars
    for j in range(1,i+1):
        print('&',end = " ")

    print()
```

```python
#✅ Pattern 4 (Right-Aligned Triangle)
n = 9
for i in range(1,n+1):

    #spaces
    for s in range(n-i):
        print(' ',end=" ")

    #stars
    for j in range(n,n-i,-1):
        print('8',end = " ")

    print()
```
