```python
#find the sum of elemnt in the list
list = [1,2,3,4,5,6,7]
sum = 0
for i in list:
    sum = sum + i
print(sum)
    

   
```

```python
#find the largest and smallest number in the list:
list = [56,78,32,90]
largest = list[0]
smallest = list[0]
for i in list:
    if i >largest:
        largest=i
    if i < smallest:
        smallest = i

print('smallest',smallest)
print('largest',largest)

```
