```python
#find the largest and smallest number in the list:
list = [56,78,32,90]
largest = list[0]
smallest = list[0]
for i in list:
    if i >largest:
        largest=i
    if i < smallest:
        smallest = i

print('smallest',smallest)
print('largest',largest)

```
