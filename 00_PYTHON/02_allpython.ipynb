```python
'''Problem 4: Write a menu-driven program -
(1) cm to ft
(2) km to miles
(3) USD to INR
(4) exit'''

while True:
    print("\n--- Conversion Menu ---")
    print("1. Centimeter to Feet")
    print("2. Kilometer to Miles")
    print("3. USD to INR")
    print("4. Exit")

    choice = int(input("Enter your choice (1-4): "))

    if choice == 1:
        cm = float(input("Enter length in cm: "))
        feet = cm / 30.48
        print("Length in Feet:", feet)

    elif choice == 2:
        km = float(input("Enter distance in km: "))
        miles = km * 0.621371
        print("Distance in Miles:", miles)

    elif choice == 3:
        usd = float(input("Enter amount in USD: "))
        inr = usd * 83   # Approximate rate
        print("Amount in INR:", inr)

    elif choice == 4:
        print("Thank you! Program exited.")
        break

    else:
        print("Invalid choice! Please try again.")
```

```python
'''fibonacci series'''
# Number of terms
n = 10

# First two numbers
a = 0
b = 1

print("Fibonacci Series:")

# Print first two terms
print(a, b, end=" ")

# Generate remaining terms
for i in range(2, n):
    c = a + b
    print(c, end=" ")
    a = b
    b = c
```
