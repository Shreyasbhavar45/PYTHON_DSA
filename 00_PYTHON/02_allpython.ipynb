```python
## Operators If + Else + Loops.
'''1.Arithemetic Operators.'''
print(5+6)

print(9*9)

print(9//8)

print(5/2)

print(5-6)

print(5%2)#gives remainder

print(5**2)#5 to the power 2
```

```python
'''2.Relational operator'''
print(4>5)
print(5<4)
print(5<=7)
print(5>=7)
print(8==7)
print(4!=10)
```

```python
'''3.Logical Operator'''
#and:- 1 and 1 = 1 otherwise = 0
print(0 and 0)

#or:- 1 or 0 = 1 otherwise = 0
print(1 or 0)

#not :- Do reverse.
print(not False)
```

```python
'''4.Bitwise Operator'''
#bitwise &
print(2&3)

#bitwise or
print(2 | 3)

#bitwise xor
print(2^3)

print(~3)
print(5>>2)
print(5<<2)
```

```python
'''5.assignment operator'''
# = 
# a = 2
a=b=c=d = 2

# a = a + 2
a += 2
b -= 2
c *= 2
d %= 8
print(a,b,c,d)
```

```python
'''6.membership operators'''
# in/not in

print("D" in "Delhi")
print("d" in 'Delhi')
print(5 in [1,3,4,5])
```

```python
#for
n = 4
for i in range(1,n+1):
    for j in range(i):
        print(j,end ='')
    print()
```
