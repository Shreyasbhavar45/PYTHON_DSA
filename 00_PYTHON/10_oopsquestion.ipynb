```python
## Student â†’ name, age
class Student:
    def __init__(self,name,age):
        self.name = name
        self.age = age

    def display(self):
        # print("Name:-",self.name)
        # print('Age:-',self.age)

        # by print in method always return it seems good
        return f'Name:-{self.name},Age:-{self.age}'

s = Student("Shreyas",19)
print(s.display())
```

```python
#- Book â†’ title, author.
class Book:
    def __init__(self,title,author):
        self.title = title
        self.author = author

    def display(self):
        return f'Title is:-{self.title},Author:-{self.author}'

B = Book("balbharti",'swamivivekanand')
print(B.display())
```

```python
##- Car â†’ brand, speed
class Car:
    
    #constructor
    def __init__(self,brand,speed):
        self.brand = brand
        self.speed = speed

    #method not a function
    def display(self):
        return f"Car Brand is :- {self.brand} , Car speed :- {self.speed}"


#use to call the class 
C = Car("Kia",190)
print(C.display())
```

```python
# ðŸ‘‰ Dog â†’ name, breed
class Dog:
    #constructor
    def __init__(self,name,bread):
        self.name = name
        self.bread = bread

    def display(self):
        return f'Dog Name is:- {self.name} , DOG bread is :- {self.bread}'
    
D = Dog("Rocky","Cheese Bread")
print(D.display())
```

```python
#
class Rectangle:
    
    def __init__(self,length,width):
        self.length = length
        self.width = width
    
    def area(self):
        return self.length*self.width
    
    def perimeter(self):
        return 2*(self.length + self.width)
    
    def display(self):
        return f'''
length : {self.length}
width : {self.width}
area : {self.area()}
perimeter : {self.perimeter()}
'''
    
R = Rectangle(3,5)
print(R.display())
```

```python
'''ðŸ‘‰ BankAccount â†’ deposit money
Requirements:
store balance
deposit increases balance
display updated balance
No withdraw yet. Only deposit.'''

class BankAccount:

    #constuctor
    def __init__(self,balance):
        self.balance = balance

    def deposit(self,amount):
        if amount > 0:

            self.balance += amount
    

    def display(self):
        return f'balance :-{self.balance}'

B = BankAccount(10000)
B.deposit(15000)
print(B.display())
```

```python
##
class mobile:

    def __init__(self,battery):
        self.battery = battery

    def use(self,amount):
        self.battery -= amount
        if self.battery < 0:
            self.battery = 0

    def display(self):
        return f'Battery :- {self.battery}'

M = mobile(90)
M.use(70)

print(M.display())
```
