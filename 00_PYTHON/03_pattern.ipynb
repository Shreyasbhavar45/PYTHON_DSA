```python
'''
Q1.print the following pattern
*

'''
n = int(input("Enter The Number:-"))

for i in range(n):
    for j in range(n):
        print('*',end=" ")
    print()
```

```python
'''
Q2.print the following pattern
**
'''
n = 1
m = 2
for i in range(n):
    for j in range(m):
        print("*",end=" ")
    print()
```

```python
#Q3. print following pattern.
'''
*
*
'''
rows = 2
coln = 1
for i in range(rows):
    for j in range(coln):
        print('*',end=" ")
    print()
```
