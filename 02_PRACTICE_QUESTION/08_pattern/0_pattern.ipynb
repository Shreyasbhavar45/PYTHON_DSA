### To print given number of *s in a rows.

```python
'''we have to print the number of star in rows'''
'''firstly we have to understand the question what we have to do'''
'''then break in a small steps'''
'''think it . then solve'''
## 5 star
def row_star(n):
    for i in range(n):#0,1,2,3,4
        print('*',end = ' ')
    print()# remeber u have to give this value always. to print the stars
row_star(5)
```

## square pattern .

```python
'''this is 1st method'''

def square_pattern(n):
    for i in range(n):
        print(' * '* n)

    print()

square_pattern(4)
```

```python
'''this is second method'''
def square_pattern(n):
    for i in range(n):
        for j in range(n):
            print('*',end =' ')
    
        print()


square_pattern(4)
```

## Rectangle pattern.

```python
def rectangle_pattern(n,m):
    for i in range(n):
        for j in range(m):
            print("*",end=" ")

        print()

rectangle_pattern(4,5)
```

# Right angled Triangle.

```python
def right_angled_triangle(n):
    for i in range(n):
        for j in range(i+1):
            print("*",end =" ")
        
        print()

right_angled_triangle(5)
```

# Inverted right angled triangle.

```python
def inverted_right_angled(n):
    for i in range(n):
        for j in range(i,n):
            print("*",end =" ")

        print()

inverted_right_angled(7)
```

## hill pattern

```python
def hill (n):
    for i in range(n):
        for j in range(i,n):
            print(" ",end=" ")
        for j in range(i):
            print("*",end=" ")
        for j in range(i+1):
            print("*",end=" ")
        print()

hill(5)            
```

```python
'''reverse hill pattern'''
def reverse_hill(n):
    for i in range(n):
        for j in range(i+n):
            print(" ",end =" ")
        for j in range(i,n-1):
            print("*",end=" ")
        for j in range(i,n):
            print("*",end=" ")
        print()

reverse_hill(5)
```
