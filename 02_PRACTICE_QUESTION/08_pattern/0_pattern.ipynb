### To print given number of *s in a rows.

```python
'''we have to print the number of star in rows'''
'''firstly we have to understand the question what we have to do'''
'''then break in a small steps'''
'''think it . then solve'''
## 5 star
def row_star(n):
    for i in range(n):#0,1,2,3,4
        print('*',end = ' ')
    print()# remeber u have to give this value always. to print the stars
row_star(5)
```

## sqaure pattern .

```python
'''this is 1st method'''

def square_pattern(n):
    for i in range(n):
        print('* '* n)

    print()

square_pattern(4)
```

```python
'''this is second method'''
def square_pattern(n):
    for i in range(n):
        for j in range(n):
            print('*',end =' ')
    
        print()


square_pattern(4)
```
