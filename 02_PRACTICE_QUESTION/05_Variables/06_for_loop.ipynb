```python
###Find the largest digit in a number (example: 58291)
n = 58291
original = 0
for digit in str(n):
    if int(digit)>original:
          original=int(digit)
print(original)
```

```python
####Find the smallest digit in a number
n = 58291
smallest = 9
for digit in str(n):
    if int(digit)< smallest:
        smallest = int(digit)
print(smallest)
```

```python
###Count how many times the digit 5 appears in a number (example: 45525)
n = 45525
count = 0
for digit in str(n):
    if int(digit)==5:
        count+=1
print(count)
```

```python
###Check whether a number contains the digit 7 (example: 78241)
n = 782417
found = False
for digit in str(n):
    if int(digit)==7:
        found = True
    
if found:
    print('yes')
else:
    print("no")
```
