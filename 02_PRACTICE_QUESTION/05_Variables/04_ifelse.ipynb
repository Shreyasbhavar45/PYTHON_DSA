## Questions of if elif and else.


```python
## 21. Vowel Check
ch = input("Enter The Alphabet")
if ch in "aieouAEIOU":
    print("Vowel")
else:
    print("Not Vowel")
```

```python
## 22. Alphabet Check
ch = "20"## check the alphabet between a to z.
if ch.isalpha():
    print("Alphabet")
else:
    print("not alphabet")
```

```python
## 23. Digit check
ch = 'alphabet'
if ch.isdigit():## check a number if there is no numbers then it not works.
    
    print("IS A DIGIT")
else:
    print("not a digit")
```

```python
##24. Special character check
ch = input("Enter a character: ")

if ch.isalpha():
    print("Alphabet")
elif ch.isdigit():
    print("Digit")
else:
    print("Special Character")

## if not ch.isalnum():
    #print("Special Character"). use to check special character
```

```python
##25. Upper/lower case
ch = 'A'
if ch.isupper():
    print("Uppercase")

else:
    print("Not a upper case letter")
```

```python
## 26.lowercase
ch ='E'
if ch.islower():
    print('lowercase')
else:
    print("Not a Lower Case")
```
