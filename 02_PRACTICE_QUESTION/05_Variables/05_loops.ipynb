### While Loop Question. [When Condition Is Not Given.]

```python
##Q1. Print numbers from 1 to 10
i = 1
while i<=10:
    print(i)
    i = i+1 #[i = i+1.. Increase the value of i by 1]## if i dont give this 
    ##condition this loop will run for infinite time.
```

```python
##Q2:Print numbers from 10 to 1 using while loop.
i = 10
while i>=1:
    print(i)
    i = i-1
```

```python
##Q3.print even numbers from 1 to 20 using while loop.
i = 2 #start
while i<=20: #condition
    print(i)
    i=i+2 #update
```

```python
##Q4: Print all odd numbers from 1 to 20 using while loop.
i = 1 #start
while i<=20: #condition
    print(i)
    i=i+2 #update
```

```python
##Q5: Print the multiplication table of a number using while loop
n = 5
i = 1 # start
while i<=10: #condition
    print(n,'*',i,'=',n*i)
    i= i+1 # update
```

```python
##Q6: Count the number of digits in a number using while loop.
n = 7567
count = 0
while n>0:
    count=count+1
    n = n//10
print(count)
```

```python
##Q7: sum of digits of number
n = 343
total=0
while n>0:
    digit =n%10
    total = total+digit
    n = n//10
print(total)
    
```
