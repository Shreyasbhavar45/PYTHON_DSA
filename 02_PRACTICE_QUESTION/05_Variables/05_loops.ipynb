‚ûï 2. Sum Programs (Python while)

- Sum of cubes of first n numbers

```python
##Sum of numbers from 1 to n

num = int(input("Enter number: "))#564
sum_digits = 0

while num > 0:
    digit = num % 10
    sum_digits += digit
    num //= 10

print(sum_digits)
```

```python
## Sum of digits of a number
n = 12342
digit = 0
while n>0:
    total= n%10
    digit+=total
    n//=10
print(digit)
```

```python
##Sum of squares of first n numbers
n = 10
i = 1
sum = 0
while i<=n:
    sum+=i**2
    i+=1
print(sum)
```

```python
###  Sum of multiples of 5 up to n
n = 100
start = 5
total = 0
while start<=n:
    total+=start
    start+=5
print(total)
```

```python
## Keep taking numbers and stop when 0 is entered, then print the sum
sum = 0
while True:
    n = int(input("Enter The number"))
    if n ==0:
        break
    sum += n
    
print(sum)
```

```python
##check a palindrome
n = int(input('enter the value:-'))
original = n
rev = 0
while n>0:
    digit=n%10
    rev = rev*10+digit
    n=n//10
print(rev)


if original==rev:
    print('palindrome')
else:
    print('not  A Palindrome')
```
