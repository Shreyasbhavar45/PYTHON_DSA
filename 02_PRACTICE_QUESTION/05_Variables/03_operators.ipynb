```python
#14. Grade calculation
marks = int(input("Enter the Marks:-"))

if marks < 0 or marks > 100:
    print("Invalid marks")

elif marks >= 90:
    print("Grade A")

elif marks >= 80:
    print("Grade B")

elif marks >= 70:
    print("Grade C")

elif marks >= 40:
    print("Grade D")

else:
    print("Fail")


```
