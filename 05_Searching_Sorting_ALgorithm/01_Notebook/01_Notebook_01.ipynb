# Linear Search:-

```python
#linear_search([1, 1, 2, 1], 1,3,4,5) should return 0
def linear_search(arr,target):
    size = len(arr)
    for index in range(0,size):
        if(arr[index]==target):
            return index
    
    return -1



my_list = [10,20,30,40,50]
target = 30
result = linear_search(my_list,target)
print(result)
```

# Question on linear search
- linear_search([3, 7, 2, 5], 2) should return 2

- linear_search([1, 1, 2, 1], 1) should return 0

- linear_search([], 5) should return -1

- linear_search([4, 2, 8], 6) should return -1

```python
## linear search ([3,7,2,5],2) should return 2
def linear_search(arr,target):
    size = len(arr)
    for index in range(0,size):
        if arr[index]==target:
            return index
        
    return -1
my_list = [2,7,2,5]
target = 5
result = linear_search(my_list,target)
print(result)
```
