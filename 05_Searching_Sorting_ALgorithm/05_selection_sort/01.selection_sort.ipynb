## Selection Sort


```python
## selcetion sort 

What it does (big idea)
- Selection sort repeatedly selects the smallest (or largest) element from the unsorted portion of the list and moves it to the end of the sorted portion.
- Think of dividing the list into two regions: a sorted prefix at the left and an unsorted suffix on the right. Each pass grows the sorted prefix by one element.

How each step works (detailed)
1. Start with i = 0 (first index). The sorted region is empty; the unsorted region is the whole array.
2. Find the index of the minimum element in the unsorted region (from index i to end). Call that minIndex.
    - Compare elements in the unsorted region one by one to update minIndex when you find a smaller value.
3. Swap the element at index i with the element at minIndex.
    - After the swap, the smallest element of the unsorted region is placed at position i.
4. Increment i by 1. Now the sorted region is from 0..i, and repeat steps 2–4 until i reaches the second-to-last element.
5. When i reaches the end, the array is fully sorted.

Illustrative example (trace)
Array: [64, 25, 12, 22, 11]

Pass 0 (i = 0):
- Unsorted region: [64, 25, 12, 22, 11]
- Find min from indices 0..4 → min = 11 at index 4
- Swap arr[0] and arr[4] → [11, 25, 12, 22, 64]

Pass 1 (i = 1):
- Unsorted region: [25, 12, 22, 64]
- Find min → 12 at index 2
- Swap arr[1] and arr[2] → [11, 12, 25, 22, 64]

Pass 2 (i = 2):
- Unsorted region: [25, 22, 64]
- Find min → 22 at index 3
- Swap arr[2] and arr[3] → [11, 12, 22, 25, 64]

Pass 3 (i = 3):
- Unsorted region: [25, 64]
- Find min → 25 (already at i)
- Swap arr[3] and arr[3] (no change) → [11, 12, 22, 25, 64]

Pass 4 (i = 4):
- Single element left → done.

Why it works
- Each pass picks the exact smallest remaining item and places it in the next sorted position. After k passes, the first k elements are the k smallest elements in sorted order. Thus after n−1 passes the whole array is sorted.

Costs and properties
- Comparisons: Always about n*(n-1)/2 comparisons (independent of input order).
- Swaps: At most n−1 swaps (one per pass). So fewer swaps than some sorts (e.g., bubble sort does many swaps).
- Time complexity: O(n^2) worst/average/best — not adaptive.
- Space complexity: O(1) extra (in-place).
- Stability: Not stable by default (equal elements can change relative order due to swaps), though a stable variant is possible with more work.
- Use cases: Simple and predictable; useful when writes/swaps are expensive and n is small. Not good for large n.

Pseudocode
- for i from 0 to n-2:
     - minIndex = i
     - for j from i+1 to n-1:
          - if arr[j] < arr[minIndex]: minIndex = j
     - swap arr[i] and arr[minIndex]

Tips to remember
- Outer loop fixes the position to fill next (i).
- Inner loop finds the minimum in the remaining part.
- Only one swap per pass (at the end), which distinguishes it from many other simple sorts.
- Visualize the sorted prefix growing left-to-right after each pass.

After reading this
- You should be able to trace selection sort on any small array by repeating "find min in unsorted, swap to front" until done.
```

## Examples


```python
def Selection_sort(arr):
    n = len(arr)  # total number of elements

    # Outer loop: har position pe correct element place karenge
    for i in range(n):

        # Step 1: maan lo current index hi minimum hai
        min_index = i  

        # Step 2: baaki array me actual smallest dhundho
        for j in range(i + 1, n):

            # agar chhota element mil gaya
            if arr[j] < arr[min_index]:
                min_index = j  # update min_index

        # Step 3: swap current element with smallest found element
        arr[i], arr[min_index] = arr[min_index], arr[i]

        # Debug print (VS Code me run karega to har step dikhega)
        print(f"Step {i+1}: {arr}")

    return arr


# Driver Code
arr = [64, 25, 12, 22, 11]

print("Original array:", arr)

result = Selection_sort(arr)

print("Sorted array:", result)
```
