## Insertion sort.


```python
#function method
def insertionSort(arr):
    for i in range(1,len(arr)):
        key = arr[i]
        j = i - 1

        while j>= 0 and key < arr[j]:
            arr[j+1]= arr[j]
            j = j - 1
        
        arr[j+1] = key

    return arr


## driver code
arr = [9,5,1,4,3]
##function calling
result =insertionSort(arr)
print("sorted array after appliying the insertion sort:-",result)
```

### Complexity & notes

- Time complexity: **O(n^2)** average and worst-case. Best-case is **O(n)** when the array is already nearly sorted (only one comparison per element).
- Space complexity: **O(1)** additional (in-place).
- Stable: **Yes** — insertion sort preserves the relative order of equal elements.
- When to use: good for small arrays or nearly-sorted data; simple to implement and efficient for short lists.


### Insertion Sort — step-by-step trace

Run the code cell below to see how insertion sort shifts elements and places the key at each step.
- `Start i=...` shows the selected key and the current array.
- `shifted:` shows the array after each shift when an element is moved right to make room for the key.
- `Placed key:` shows the array after inserting the key into its correct position.

This trace helps you visualise exactly how elements move so you can understand the algorithm clearly even after some time.

```python
def insertionSort_trace(arr):
    # work on a copy so the caller's list isn't modified
    arr = list(arr)
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        print(f"Start i={i}, key={key}, arr={arr}")
        while j >= 0 and key < arr[j]:
            arr[j+1] = arr[j]
            j -= 1
            print(f"  shifted: arr={arr}, j={j}")
        arr[j+1] = key
        print(f"Placed key: arr={arr}\n")
    return arr

# Demo: run the trace on the example array
print('Input: [9,5,1,4,3,90,80]')
insertionSort_trace([9,5,1,4,3])
```
