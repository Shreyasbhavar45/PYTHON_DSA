```python
##Binary Search
def BinarySearch(arr,target):
    start = 0
    end = len(arr)-1

    #binarysearch implementation
    while start <= end :
        #to find the middle index of the arr
        mid = start + (end-start)//2

        if arr[mid] == target:
            return mid
        
        elif arr[mid] < target:
            start = mid + 1

        else:
            end = mid - 1

    return -1



##driver code
arr = [10,20,30,40,50,60,70,80]
target = 50
#function calling
result = BinarySearch(arr,target)
print('The element is found at index is :-',result)
```

```python
## search element at index
def Searchinsert(arr,target):
    start = 0
    end = len(arr)-1

    #binary search implementation
    while start <= end:

        #to find the mid element
        mid = start + (end - start)//2

        if arr[mid] == target:
            return mid
        
        elif arr[mid] < target:
            start = mid + 1
        
        else:#arr[mid]> target:
            end = mid - 1
    
    return start

##drive code
arr = [10,20,30,40,50,60,70,80]
target = 35
#function calling
result = Searchinsert(arr,target)
print('The element is found at index is :-',result)

'''Since the array is sorted, I use binary search. 
 I compare the middle element with the target and adjust the search range accordingly. 
 If the target is found, I return its index. Otherwise, 
 when the loop ends, the start pointer gives the correct insert position. 
 The time complexity is O(log n)'''
```

```python
def countNegatives(arr):

    rows = len(arr)
    cols = len(arr[0])

    r = 0
    c = cols - 1
    count = 0

    while r < rows and c >= 0:

        if arr[r][c] < 0:
            count += rows - r
            c -= 1
        else:
            r += 1

    return count


# Given input from question
arr = [[4,3,2,1],[3,2,1,1],[1,1,-1,-2],[-1,-1,-2,-3]]
result = countNegatives(arr)
print(result)






```
