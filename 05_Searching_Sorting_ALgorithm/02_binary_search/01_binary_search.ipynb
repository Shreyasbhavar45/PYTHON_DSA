## Binary Search
- is a very fast searching algorithm used to find an element inside a sorted list or array.

## how it works
- (1)- check the middle element.
- (2)- if it is equal to the target ->found
- (3)if the target is smaller -> search in the left half.
- (4)if the target is bigger ->search in the right half.
- (5)repeat until the range become empty.


```python
## Binary Search Algorithm without using recursion(iterative)
## function definition
def binarySearch(arr, x, i, j):
    while i <= j:
        ## lower bound while division
        mid = i + (j-i)//2
        if arr[mid] == x:
            return mid
        elif arr[mid] < x:
            ## right side of the mid
            ## seach space - mid+1 to j
            i = mid + 1
            
        else: ## arr[mid] > x:
            ## left side of the mid
            ## search space - i to mid-1
            j = mid - 1
    ## searching element is not present in an array
    return -1

## Driver code
arr = [20, 30, 40, 50, 60, 70, 80, 90]
x = 70
i = 0#left side 
j = len(arr) - 1#right side
## function calling
result = binarySearch(arr, x, i, j)
print("Searching element is present at the location:",result)
```

```python
## Binary Search Algorithm using recursion
## function definition
def binarySearch(arr, x, i, j):
    while i <= j:
        mid = i + (j-i)//2
        if arr[mid] == x:
            return mid
        elif arr[mid] < x:
            ## right side of the mid
            ## seach space - mid+1 to j
            ## recursion - calling the same function again inside the 
            ## method definition
            return binarySearch(arr, x, mid+1, j)
        else:
            ## left side of the mid
            ## search space - i to mid-1
            return binarySearch(arr, x, i, mid-1)
    ## searching element is not present in an array
    return -1

## Driver code
arr = [20, 30, 40, 50, 60, 70, 80, 90]
x = 80
i = 0
j = len(arr) - 1
## function calling
result = binarySearch(arr, x, i, j)
print("Searching element is present at the location:",result)
```

```python
##Best time to buy and sell the stocks.
##time complexity :- O(n)
##function Definition 
def FindMaxProfit(price):
    #initialization.
    minprice = float('inf')#infinite value,Taaki first element automatically minimum ban jaye.
    maxprofit = 0

    #Ye loop har price ko ek baar check karta hai.
    for i in range(len(price)):

                # -------- Day 1 --------
        # price = 7
        # 7 < inf → True
        # isliye minprice = 7 ho jayega
        # profit calculate nahi hoga kyuki abhi buy price hi set hua hai

        # -------- Day 2 --------
        # price = 4
        # 4 < 7 → True
        # minprice update hoga
        # minprice = 4

        # -------- Day 3 --------
        # price = 5
        # 5 < 4 → False
        # profit check karenge
        # profit = 5 - 4 = 1
        # maxprofit = 1

        # -------- Day 4 --------
        # price = 3
        # 3 < 4 → True
        # minprice update
        # minprice = 3

        # -------- Day 5 --------
        # price = 6
        # 6 < 3 → False
        # profit = 6 - 3 = 3
        # maxprofit = 3

        # -------- Day 6 --------
        # price = 16
        # 16 < 3 → False
        # profit = 16 - 3 = 13
        # maxprofit = 13

        if price[i]< minprice:
            minprice = price[i]
        
        elif price[i] - minprice > maxprofit:
            maxprofit = price[i]-minprice
        
    return maxprofit



##driver code
price = [7,4,5,3,6,16]
maxprofit_value = FindMaxProfit(price)
print('the max profit of buying and selling stock is:-',maxprofit_value)

'''We iterate through the array once and keep track of the minimum price seen so far.
For each day we calculate the profit if we sell today.
We update the maximum profit accordingly.
Since we traverse the array only once, the time complexity is O(n) and space complexity is O(1).'''
```

```python
## function definition
## time complexity: O(log(m*n))
def Search2Darray(arr,target):##m*n
    ##number of rows
    m = len(arr)
    if m == 0:
        return False
    ##Number of Columns
    n = len(arr[0])
    left,right = 0 , m*n-1
    ##binary search implementation.
    while left <= right:
        mid = left + (right - left)//2
        #row_number -> mid//n
        #column_number -> mid%n
        mid_element = arr[mid//n][mid%n]
        if target == mid_element:
            return True
        elif target < mid_element:
            right = mid - 1
        else:#target > mid_element:
            left = mid + 1
    return False




##driver code
arr = [[1,3,5,7],[10,11,16,20],[23,30,34,60]]
target = 1
result = Search2Darray(arr,target)
print(result)
```

```python
# Problem:
# Search a target value in a 2D matrix.
# Conditions of matrix:
# 1. Each row is sorted in increasing order.
# 2. First element of every row is greater than last element of previous row.

# Because of these properties, the matrix can be treated like a single sorted 1D array.
# Example:
# Matrix:
# [1, 3, 5, 7]
# [10,11,16,20]
# [23,30,34,60]

# Flattened (conceptually):
# [1,3,5,7,10,11,16,20,23,30,34,60]

# Therefore we can apply Binary Search.


# Function Definition
# Time Complexity: O(log(m*n))
# m = number of rows
# n = number of columns

def Search2Darray(arr, target):

    # Step 1: Find number of rows
    m = len(arr)

    # If matrix is empty
    if m == 0:
        return False

    # Step 2: Find number of columns
    n = len(arr[0])

    # Step 3: Binary Search range
    # Since total elements = m*n
    left = 0
    right = m*n - 1

    # Step 4: Binary Search Loop
    while left <= right:

        # Middle index
        mid = left + (right - left) // 2

        # Convert 1D index to 2D index
        # Row number
        row = mid // n

        # Column number
        col = mid % n

        # Access element from matrix
        mid_element = arr[row][col]

        # Step 5: Compare with target
        if target == mid_element:
            return True

        # If target is smaller → search left side
        elif target < mid_element:
            right = mid - 1

        # If target is larger → search right side
        else:
            left = mid + 1

    # If element not found
    return False


# Driver Code
arr = [[1,3,5,7],[10,11,16,20],[23,30,34,60]]

target = 1

result = Search2Darray(arr, target)

print(result)
```

## Binary Search Cheat Sheet

This short guide helps you reframe problems so binary search applies. Follow these steps:

1. Identify a sorted or monotonic domain (array indices, value thresholds, feasibility).
2. Define a monotonic predicate P(mid) that is True for all indices on one side of the answer and False on the other.
3. Choose an invariant and bounds style (inclusive [L,R] or half-open [L,R)). Half-open is less error-prone.
4. Use mid = L + (R-L)//2 and update bounds preserving the invariant.
5. Test edge cases: empty, single element, all < target, all > target, duplicates.

Common templates are below — copy them into a code cell to use directly.

```python
# Template A: Classic search (returns index or -1)
def binary_search(arr, target):
    L, R = 0, len(arr) - 1
    while L <= R:
        mid = L + (R - L) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            L = mid + 1
        else:
            R = mid - 1
    return -1

# Template B: lower_bound (first index with arr[idx] >= target)
def lower_bound(arr, target):
    L, R = 0, len(arr)
    while L < R:
        mid = L + (R - L) // 2
        if arr[mid] < target:
            L = mid + 1
        else:
            R = mid
    return L  # may be len(arr) if not found

# Template C: Generic predicate search (first True)
def find_first_true(low, high, predicate):
    # searches integers in [low, high], returns first x where predicate(x) is True, or high+1 if none
    L, R = low, high + 1
    while L < R:
        mid = L + (R - L) // 2
        if predicate(mid):
            R = mid
        else:
            L = mid + 1
    return L

# Template D: 2D matrix search (flatten mapping)
def search_2d_matrix(matrix, target):
    if not matrix or not matrix[0]:
        return False
    m, n = len(matrix), len(matrix[0])
    L, R = 0, m * n - 1
    while L <= R:
        mid = L + (R - L) // 2
        val = matrix[mid // n][mid % n]
        if val == target:
            return True
        elif val < target:
            L = mid + 1
        else:
            R = mid - 1
    return False

# Quick tests
print('binary_search:', binary_search([1,2,3,4,5], 3))
print('lower_bound (first >=4):', lower_bound([1,2,4,4,5], 4))
print('find_first_true demo (first x >= 7):', find_first_true(0, 20, lambda x: x >= 7))
print('search_2d_matrix:', search_2d_matrix([[1,3,5],[7,9,11]], 9))
```
