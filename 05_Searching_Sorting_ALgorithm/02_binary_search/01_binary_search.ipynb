## Binary Search
- is a very fast searching algorithm used to find an element inside a sorted list or array.

## how it works
- (1)- check the middle element.
- (2)- if it is equal to the target ->found
- (3)if the target is smaller -> search in the left half.
- (4)if the target is bigger ->search in the right half.
- (5)repeat until the range become empty.


```python
## Binary Search Algorithm without using recursion(iterative)
## function definition
def binarySearch(arr, x, i, j):
    while i <= j:
        ## lower bound while division
        mid = i + (j-i)//2
        if arr[mid] == x:
            return mid
        elif arr[mid] < x:
            ## right side of the mid
            ## seach space - mid+1 to j
            i = mid + 1
            
        else: ## arr[mid] > x:
            ## left side of the mid
            ## search space - i to mid-1
            j = mid - 1
    ## searching element is not present in an array
    return -1

## Driver code
arr = [20, 30, 40, 50, 60, 70, 80, 90]
x = 70
i = 0#left side 
j = len(arr) - 1#right side
## function calling
result = binarySearch(arr, x, i, j)
print("Searching element is present at the location:",result)
```

```python
## Binary Search Algorithm using recursion
## function definition
def binarySearch(arr, x, i, j):
    while i <= j:
        mid = i + (j-i)//2
        if arr[mid] == x:
            return mid
        elif arr[mid] < x:
            ## right side of the mid
            ## seach space - mid+1 to j
            ## recursion - calling the same function again inside the 
            ## method definition
            return binarySearch(arr, x, mid+1, j)
        else:
            ## left side of the mid
            ## search space - i to mid-1
            return binarySearch(arr, x, i, mid-1)
    ## searching element is not present in an array
    return -1

## Driver code
arr = [20, 30, 40, 50, 60, 70, 80, 90]
x = 80
i = 0
j = len(arr) - 1
## function calling
result = binarySearch(arr, x, i, j)
print("Searching element is present at the location:",result)
```

```python
## Two pointer Approach question
## time complexity : O(n)
def Two_sum(arr,target_sum):
    left = 0
    right = len(arr)-1

    while left<right:

        #'''basically how it works that '''for my understanding
# in step 1:- when left = 20 and right = 240 it = 260. in this condition 260 > 210.so we have to decrease it from right side.
# in step 2:- when left = 20 and right = 120 it = 140. in this condition 140 < 210. so we have to increment the value of left by 1.now the value of left is 40.
# in step 3:- when left = 40 and right = 120 it = 160. in this condition 160 < 210. so we have to increment the value of left by 1. now the value of left is 60.
# in step 4:- when left = 60 and right = 120 it = 180. in this condition 180 < 210. so we have to increment the value of left by 1. now the value of left side is 80.
# in step 5:- when left = 80 and right = 120 it = 200. in this condition 200 < 210. so we have to increment the value of left by 1. now the value of left side is 90.
# in step 6:- when left = 90 and right = 120 it = 210. in this condition 210 = 210. so this time the conditon is satisfy and target_value is find. and the answer is (4,5).
        if arr[left]+arr[right]==target_sum:
            return (left,right)
        

        elif arr[left]+arr[right]<target_sum:
            #agar target_sum chota hai toh left ko aage badao(increament krao toh ek se value bad jaegii)
            ##Sum kam hai → bada number chahiye → left++
            left = left + 1
        
        else:#arr[left]+arr[right]>target_sum:
            #agar target_sum bada hai toh right ko piche lao.
            ##Sum bada hai → chhota number chahiye → right--
            right = right - 1
    
    return -1

## Driver code 
## Array is in a sorted way.
arr = [20,40,60,80,90,120,240]
target_sum = 210
result = Two_sum(arr,target_sum)
print(result)


''' 
Since the array is sorted, we use the two pointer approach.
One pointer starts from the beginning and the other from the end.
If the sum is smaller than the target we move the left pointer forward,
if the sum is larger we move the right pointer backward.
This allows us to find the pair in O(n) time instead of O(n²).
'''
```

```python
##Best time to buy and sell the stocks.
##time complexity :- O(n)
##function Definition 
def FindMaxProfit(price):
    #initialization.
    minprice = float('inf')#infinite value,Taaki first element automatically minimum ban jaye.
    maxprofit = 0

    #Ye loop har price ko ek baar check karta hai.
    for i in range(len(price)):

                # -------- Day 1 --------
        # price = 7
        # 7 < inf → True
        # isliye minprice = 7 ho jayega
        # profit calculate nahi hoga kyuki abhi buy price hi set hua hai

        # -------- Day 2 --------
        # price = 4
        # 4 < 7 → True
        # minprice update hoga
        # minprice = 4

        # -------- Day 3 --------
        # price = 5
        # 5 < 4 → False
        # profit check karenge
        # profit = 5 - 4 = 1
        # maxprofit = 1

        # -------- Day 4 --------
        # price = 3
        # 3 < 4 → True
        # minprice update
        # minprice = 3

        # -------- Day 5 --------
        # price = 6
        # 6 < 3 → False
        # profit = 6 - 3 = 3
        # maxprofit = 3

        # -------- Day 6 --------
        # price = 16
        # 16 < 3 → False
        # profit = 16 - 3 = 13
        # maxprofit = 13

        if price[i]< minprice:
            minprice = price[i]
        
        elif price[i] - minprice > maxprofit:
            maxprofit = price[i]-minprice
        
    return maxprofit



##driver code
price = [7,4,5,3,6,16]
maxprofit_value = FindMaxProfit(price)
print('the max profit of buying and selling stock is:-',maxprofit_value)

'''We iterate through the array once and keep track of the minimum price seen so far.
For each day we calculate the profit if we sell today.
We update the maximum profit accordingly.
Since we traverse the array only once, the time complexity is O(n) and space complexity is O(1).'''
```

```python
## function definition
## time complexity: O(log(m*n))
def search2DArray(arr, target):
    ## number of rows
    m = len(arr)
    if m == 0:
        return False
    ## number of columns
    n = len(arr[0])
    left, right = 0, m*n-1
    ## binary search implementation
    while left <= right:
        mid = left + (right - left)//2
        ## row_number -> mid//n
        ## column_number -> mid%n
        mid_element = arr[mid//n][mid%n]
        if target == mid_element:
            return True
        elif target < mid_element:
            right = mid - 1
        else:
            left = mid + 1
    return False


## Driver code
arr = [[1,3,5,7],[10,11,16,20], [23,30,34,60]]
target = 60
result = search2DArray(arr, target)
print(result)
```
