```python
##Bubble sort Implementation.

def Bubble_sort(arr):
    start = 0
    end = len(arr)

    for i in range(end):
        for j in range(end-i-1):
            if arr[j] > arr[j+1]:
                #swap of the element.
                arr[j],arr[j+1] = arr[j+1],arr[j]

    return arr
            



##Driver Code
arr = [70,20,50,30,90,5,15]
#function calling
result = Bubble_sort(arr)
print("array after appliying bubble sort",result)
```
