```python
def TernarySearch(arr, start, end, target):
    if start > end:
        return -1

    mid1 = start + (end - start) // 3
    mid2 = end - (end - start) // 3

    if arr[mid1] == target:
        return mid1

    if arr[mid2] == target:
        return mid2

    # First part
    if target < arr[mid1]:
        return TernarySearch(arr, start, mid1 - 1, target)

    # Third part
    elif target > arr[mid2]:
        return TernarySearch(arr, mid2 + 1, end, target)

    # Middle part
    else:
        return TernarySearch(arr, mid1 + 1, mid2 - 1, target)


# Driver code
arr = [10, 20, 30, 40, 50, 60, 70]
target = 60

result = TernarySearch(arr, 0, len(arr) - 1, target)

print("The value is found at index:", result)
```
