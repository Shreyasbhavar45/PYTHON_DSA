```python
##function definition
def TernarySearch(arr,target):
    start = 0
    end = len(arr)-1

    while start <= end:
        mid1 = start + (end-start)//3
        mid2 = end-(end-start)//3

        if  arr[mid1]==target:
            return mid1
            
        elif arr[mid2]== target:
            return mid2
            
        elif target < arr[mid1]:
            end = mid1-1
            
        elif target > arr[mid2]:
            start = mid2 + 1

        else:
            mid1+1,mid2-1

    return -1
##Driver code
arr = [10,20,30,40,50,60,70]
target = 189
result = TernarySearch(arr,target)
print("The value is found in ternary search is :-",result)
```
