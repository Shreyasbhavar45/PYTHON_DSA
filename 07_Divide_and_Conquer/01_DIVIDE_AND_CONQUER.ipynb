```python
##method definition of max and min function()
def findMaxandMin(arr,i,j):
    ##small problem
    ##singlle element condition
    if i == j:
        max_val = arr[i]
        min_val = arr[i]
    
    ## two element condition
    elif i == j - 1:
        if arr[i]<arr[j]:
            max_val = arr[j]
            min_val = arr[i]
        
        else:
            max_val = arr[i]
            min_val = arr[j]
    
    
    
    ##big problem -> n > 2
    else:
        ##Divide and Conquer Approach
        ## 1. Divide
        mid = i + (j-i)//2
        ## 2. recusrion -> conquer
        max_1 , min_1 = findMaxandMin(arr,i,mid)
        max_2 , min_2 = findMaxandMin(arr,mid+1,j)

        #3.combine
        #To find the maxima
        if max_1 < max_2:
            max_val = max_2
        
        else:
            max_val = max_1
        
        #To find the minima

        if min_1 < min_2:
            min_val = min_1
        
        else:
            min_val = min_2
    
    return max_val , min_val


##driver code
arr = [20,39,45,65,21,44,89,92]
## i indicates the starting index 
i = 0
## j indicates the ending index
j = len(arr)-1
##function calling
max_val , min_val = findMaxandMin(arr,i,j)
print('Maximum and Minimum Value in the array is:-',max_val,min_val)
```

## __POWER ELEMENT QUESTION__

```python
def findPowerElement(a,n):
    ##small Problem
    if n == 1:
        return a
    
    elif n == 0:
        return 1
    
    else:
        ##big problem
        ## 1.Divide
        mid = n//2

        ##conquer
        b = findPowerElement(a, mid)

        ##3. combine part
        result = b * b
        if n % 2 == 0:
            return result
        return result * a
        

##driver code
a = 2
n = 16
##function Calling
result = findPowerElement(a,n)
print("The Power of an Element is :-",result)
```

## __MERGE SORT__

```python
def mergeProcedure(arr,i,mid,j):
    #n1 -> no . of . element in the left subarray(i,mid)
    n1 = mid - i + 1
    #n2 -> no of element in the right subarray ( mid + 1,j)
    n2 = j - mid

    ##initailization of left and right subarray:
    left_subarray = [0] * n1
    right_subarray = [0] * n2

    #copy the element from the array to the subarray
    for m in range(n1):
        left_subarray[m] = arr[i+m]

    for n in range(n2):
        right_subarray[n] = arr[mid+1+n]

    p = 0
    q = 0
    k = i

    ##returning a sorted array:
    while p < n1 and q < n2:
        if left_subarray[p] <= right_subarray[q]:
            arr[k] = left_subarray[p]
            p+=1
        else:
            arr[k] = right_subarray[q]
            q+=1
        k+=1
    
    ##copy the entire elements from the left subarray
    while p < n1:
        arr[k] = left_subarray[p]
        p+=1
        k+=1
    
    #copy the entire element from right_subaaray
    while q < n2:
        arr[k] = right_subarray[q]
        q+=1
        k+=1


##function definition of merge sort
##approach -> Divide and Conquer
##Recurrence Relation -> 2T(N/2)+N
##Time Complexity ->O(NlogN)

def mergeSort(arr,i,j):
    if i < j:
        #divide
        mid = i + (j - i)//2
        ##conquer
        ##Recursive call -> left subtree
        mergeSort(arr,i,mid)
        ##Recursive call -> right subtree
        mergeSort(arr,mid+1,j)
        ##combine -> MergeProcedure (function calling)
        mergeProcedure(arr,i,mid,j)
    return arr

##driver code
arr = [50,70,65,13,80,62,98,27]
##starting index
i = 0
##ending index 
j = len(arr)-1
##function calling
result = mergeSort(arr,i,j)
print("After Merge the arr we got -",result)
```
