```python
##method definition of max and min function()
def findMaxandMin(arr,i,j):
    ##small problem
    ##singlle element condition
    if i == j:
        max_val = arr[i]
        min_val = arr[i]
    
    ## two element condition
    elif i == j - 1:
        if arr[i]<arr[j]:
            max_val = arr[j]
            min_val = arr[i]
        
        else:
            max_val = arr[i]
            min_val = arr[j]
    
    
    
    ##big problem -> n > 2
    else:
        ##Divide and Conquer Approach
        ## 1. Divide
        mid = i + (j-i)//2
        ## 2. recusrion -> conquer
        max_1 , min_1 = findMaxandMin(arr,i,mid)
        max_2 , min_2 = findMaxandMin(arr,mid+1,j)

        #3.combine
        #To find the maxima
        if max_1 < max_2:
            max_val = max_2
        
        else:
            max_val = max_1
        
        #To find the minima

        if min_1 < min_2:
            min_val = min_1
        
        else:
            min_val = min_2
    
    return max_val , min_val


##driver code
arr = [20,39,45,65,21,44,89,92]
## i indicates the starting index 
i = 0
## j indicates the ending index
j = len(arr)-1
##function calling
max_val , min_val = findMaxandMin(arr,i,j)
print('Maximum and Minimum Value in the array is:-',max_val,min_val)
```

```python
def findPowerElement(a,n):
    ##small Problem
    if n == 1:
        return a
    
    elif n == 0:
        return 1
    
    else:
        ##big problem
        ## 1.Divide
        mid = n//2

        ##conquer
        b = findPowerElement(a, mid)

        ##3. combine part
        result = b * b
        if n % 2 == 0:
            return result
        return result * a
        

##driver code
a = 2
n = 16
##function Calling
result = findPowerElement(a,n)
print("The Power of an Element is :-",result)
```
