# __STRINGS QUESTION__

```python
#Q1.VAlID PALINDROME
#brute force approach
def isPalindrome(s):
    new_str = ''
    for ch in s:
        if ch.isalnum():
            new_str = new_str + ch.lower()
    
    
    return new_str == new_str[::-1]
        



#driver code 
s = 'a man , a plan , a canal:Panama'
result = isPalindrome(s)
print(result)
```

```python
##optimal approach
#two pointer approach
# time complexity = O(n)
#space complexity = O(1)
def isPalindrome(s):
    left = 0
    right = len(s)-1

    while left < right:
        if not s[left].isalnum():
            left = left + 1
            continue
        
        elif not s[right].isalnum():
            right = right - 1

        elif s[left].lower() != s[right].lower():
            return False

        else:
            left +=1
            right -=1
    return True


##drivercode
s = 'a man , a plan , a canal:Panama'
result = isPalindrome(s)
print(result)
```

## __Longest Common Prefix__

```python
def longestCommonPrefix(strs):
    # Step 1: edge case
    if not strs:
        return ""

    # Step 2: find minimum length
    min_len = len(strs[0])
    for s in strs:
        if len(s) < min_len:
            min_len = len(s)

    # Step 3: compare characters
    for i in range(min_len):
        for s in strs:
            if s[i] != strs[0][i]:
                return strs[0][0:i]

    # Step 4: if no mismatch
    return strs[0][0:min_len]

strs = ['flower','float','flore']
result = longestCommonPrefix(strs)
print(result)
```

```python
##optimal appraoch
def longestCommonPrefix(strs):
    # Edge case
    if not strs:
        return ""

    # Step 1: sort the list
    strs.sort()

    # Step 2: take first and last string
    first = strs[0]
    last = strs[-1]

    # Step 3: compare characters
    i = 0
    while i < len(first) and i < len(last):
        if first[i] != last[i]:
            break
        i += 1

    # Step 4: return prefix
    return first[0:i]

strs = ['flower','float','flore']
result = longestCommonPrefix(strs)
print(result)
```

## __Q3.REVERSE WORDS__

```python
def reverseWords(s):
    
    r = s.split()
    
    f = r[::-1]
    return " ".join(f)

s = "  nikhil sudhakar sarve "
result = reverseWords(s)
print(result)
```

```python
def reverseEachWord(s):
    words = s.split()
    res = []

    for w in words:
        res.append(w[::-1])

    return " ".join(res)

s = "  nikhil sudhakar sarve "
result = reverseEachWord(s)
print(result)
```

## __Q4.ZIGZAG CONVERSION__

```python
def convert(s, numRows):
    if numRows == 1:
        return s

    rows = [""] * numRows
    cur_row = 0
    going_down = False

    for ch in s:
        rows[cur_row] += ch

        if cur_row == 0 or cur_row == numRows - 1:
            going_down = not going_down

        if going_down:
            cur_row += 1
        else:
            cur_row -= 1

    return "".join(rows)


# 🔽 User input
s = "VIRATKOHLI"
numRows = 3

# 🔽 Function call
result = convert(s, numRows)

# 🔽 Output
print("Zigzag Conversion:", result)
```
