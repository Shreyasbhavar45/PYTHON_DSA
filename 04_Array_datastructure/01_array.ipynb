```python
## Search for an element "15" and if it's present in an array, return the index of that element.
## suppose the searching element is not present in array, return -1
## Time complexity : O(n)
## space Complexity : O(1)
## function definition
def linear_Search(arr,x):
    for i in range(len(arr)):
        if arr[i]==x:
            return i
        
    return -1



## Driver Code
arr = [2,1,8,9,12,15,11,19,49,60,30,10]
x = 15
## function calling
result = linear_Search(arr,x)
print('searching element is present at the index',result)
```
