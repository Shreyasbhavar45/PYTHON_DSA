```python
# ================================
# ARRAY PREREQUISITES (BASICS)
# ================================

# 1. ARRAY TRAVERSAL
def traverse_array(arr):
    print("Traversal:")
    for i in range(len(arr)):
        print(f"Index {i} -> {arr[i]}")
    print()


# 2. FIND MAX ELEMENT
def find_max(arr):
    max_val = arr[0]
    for num in arr:
        if num > max_val:
            max_val = num
    return max_val


# 3. SUM OF ARRAY
def array_sum(arr):
    total = 0
    for num in arr:
        total += num
    return total


# 4. REVERSE ARRAY (TWO POINTER)
def reverse_array(arr):
    i = 0
    j = len(arr) - 1

    while i < j:
        arr[i], arr[j] = arr[j], arr[i]  # swap
        i += 1
        j -= 1

    return arr


# 5. MOVE ZEROES (TWO POINTER)
def move_zeroes(arr):
    j = 0  # position for non-zero

    for i in range(len(arr)):
        if arr[i] != 0:
            arr[j], arr[i] = arr[i], arr[j]
            j += 1

    return arr


# 6. MISSING NUMBER (SUM FORMULA)
def missing_number(arr):
    n = len(arr)
    expected = n * (n + 1) // 2
    actual = sum(arr)
    return expected - actual


# 7. PALINDROME CHECK (TWO POINTER)
def is_palindrome(arr):
    i = 0
    j = len(arr) - 1

    while i < j:
        if arr[i] != arr[j]:
            return False
        i += 1
        j -= 1

    return True


# ================================
# DRIVER CODE (TEST EVERYTHING)
# ================================
if __name__ == "__main__":
    arr = [1, 2, 0, 4, 0, 5]

    traverse_array(arr)

    print("Max Element:", find_max(arr))
    print("Sum:", array_sum(arr))
    print("Reverse:", reverse_array(arr.copy()))
    print("Move Zeroes:", move_zeroes(arr.copy()))
    print("Missing Number:", missing_number([0,1,3,4]))
    print("Is Palindrome:", is_palindrome([1,2,2,1]))
```

## __Q1.Maximum Element in List__



```python
##Approach 1 :-Brute Force Approach. Compare 1 ,1 element with each other.
##Function Calling
def MaxElement(list1):
    n = len(list1)#6

    for i in range(n):#0 to 5
        ismax = True

        for j in range(n):#0 to 5
            #it means when i = 0 and j = 0 the also continue the statement.
            # i = 1 and j = 1
            # i = 2 And j = 2
            # i = 3 and j = 3
            # i = 4 and j = 4
            # i = 5 and j = 5
            if i == j:
                continue

            if list1[j]>list1[i]:
                ismax = False
                break

        if ismax:
            return list1[i]
        

##Driver code
list1 = [3,5,2,9,20,6]
#function Calling
Result = MaxElement(list1)
print("Maximum Element In the list is:-",Result)
```

```python
##Approach 2:- Optimal Approach. Assume the max element
#function Definition
def MaxElement(arr):
    n = len(arr)
    max = arr[0] # declare the index 0 element as max.
    for i in range(n):#0 to 5

        ##if arr[i]> max element that we store . update max = new arr[i]
        if arr[i]>max:
            max = arr[i]

    return max


##driver code
arr = [3,5,2,9,20,6]
#function Calling
Result = MaxElement(arr)
print("Maximum Element In the list is:-",Result)
```

## __Q3.Palindrome List__


```python
##Approach
##function definition.
def is_palindrome(arr):
    left = 0
    right = len(arr)-1

    while left<right:
        if arr[left]!=arr[right]:
            return False
        
        ## this is a pointer approach check the element from left side and right side.
        left = left+1
        right = right -1
    
    return True
        


##Driver code
arr = [1,2,3,2,1]
result = is_palindrome(arr)
print('Is the List palindrome:-',result)
```

##
- left = 0 (start), right = last index

- Swap start and end elements
- Move pointers towards center

- Repeat until left < right

- Pointers cross → stop

```python
# 🔹 Function: Reverse a list in-place using Two Pointer Approach
#Time Complexity = O(n) Each element visited once.
#Space Complexity = O(1) No extra space (in-place).
def reverse_list(arr):

    # Step 1: Initialize two pointers
    left = 0                    # starting index
    right = len(arr) - 1        # last index

    # Step 2: Run loop until left < right
    # Reason: jab pointers cross ho jayenge tab swapping complete ho chuki hogi
    while left < right:

        # Step 3: Swap elements
        # Example: arr[left] ↔ arr[right]
        # [6,4,8,2,9,0] → swap 6 and 0 → [0,4,8,2,9,6]
        arr[left], arr[right] = arr[right], arr[left]

        # Step 4:  Two Move pointers # pointer approach
        left = left + 1         # move forward
        right = right - 1       # move backward

    # Step 5: Return reversed list
    return arr


# 🔹 Driver Code
arr = [6, 4, 8, 2, 9, 0]

# Function Calling
result = reverse_list(arr)

# Output
print("The reverse list is :-", result)
```

## __Q5.Rotate List__
- Left  = full reverse last me
- Right = full reverse pehle
- soln:-def rotate_right(arr, d):

-    reverse(arr, 0, n-1)
-    reverse(arr, 0, d-1)
-    reverse(arr, d, n-1)

```python
# 🔹 Helper function: reverse part of array
def reverse(arr, start, end):
    while start < end:
        arr[start], arr[end] = arr[end], arr[start]
        start += 1
        end -= 1


# 🔹 Main function: rotate left by D positions
def rotate_left(arr, d):
    n = len(arr)

    # Edge case: if d > n
    d = d % n#16%5= 1 then the output of arr should be [2,3,4,5,1]

    # Step 1: Reverse first d elements
    reverse(arr, 0, d - 1)#index 0 or 1

    # Step 2: Reverse remaining elements
    reverse(arr, d, n - 1)#index [2 ,3, 4]

    # Step 3: Reverse whole array
    reverse(arr, 0, n - 1)

    return arr


# 🔹 Driver Code
arr = [1, 2, 3, 4, 5]
d = 16

result = rotate_left(arr, d)
print("Rotated Array:", result)
```

## __Q6.plus one.__
- approach solve for this by me -
- Step 1:
- Right (last index) se start karunga

- Step 2:
- Har digit check karunga

- Step 3:
- Agar current digit < 9 hai:
- usme +1 karunga
- aur wahi stop (return)

- Step 4:
- Agar current digit = 9 hai:
- usko 0 banaunga
- carry next (left) element me jayega

- Step 5:
- Loop continue karega jab tak:
- digit < 9 na mil jaye

- Step 6 (Special Case):
- Agar sab digits 9 ho (jaise [9,9,9]):
- new array banaunga
- first element = 1
- baaki sab = 0

- Final Output:
- Updated array return karunga

```python
#Approach
#method definition
def plus_one(arr):
    n = len(arr)
    right = len(arr)-1
    for i in range(right,-1,-1):
        if arr[i]==9:
            arr[i]=0

            continue

        elif arr[i] < 9:
            arr[i]+=1
            return arr
        
    return [1] + [0] * n

##driver code
arr = [1,2,3,4]
#function calling
result = plus_one(arr)
print(result)
```

```python
## good versionnn
# 🔹 Problem:
# Given an array of digits representing a number,
# add +1 to the number and return updated array.

# 🔹 Example:
# [1,2,3] → [1,2,4]
# [1,2,9] → [1,3,0]
# [9,9,9] → [1,0,0,0]


# 🔹 Approach (Important 🔥)

# Step 1:
# Right (last index) se start karna hai

# Step 2:
# Har digit check karna hai

# Step 3:
# Agar digit < 9 hai:
# → usme +1 karna hai
# → aur wahi stop (return array)

# Step 4:
# Agar digit == 9 hai:
# → usko 0 bana dena hai
# → carry next (left) element me jayega

# Step 5:
# Loop continue karega jab tak:
# → digit < 9 na mil jaye

# Step 6 (Special Case):
# Agar loop khatam ho gaya:
# → iska matlab sab digits 9 the
# → new array banega: [1] + [0]*n


# 🔹 Code (Final Correct Version)

def plus_one(arr):
    n = len(arr)

    for i in range(n-1, -1, -1):

        # Case 1: 9 hai → carry
        if arr[i] == 9:
            arr[i] = 0

        # Case 2: < 9 → +1 and stop
        else:
            arr[i] += 1
            return arr  # 🔥 yahi game changer hai turant arr ko stop krta hai

    # Case 3: All 9 (loop finish ho gaya)
    return [1] + [0] * n


# 🔹 Driver Code
arr = [1,2,3,4]
result = plus_one(arr)
print(result)


# 🔹 Complexity

# Time Complexity  = O(n)
# Space Complexity = O(1)


# 🔹 One-line Logic

# "Right se start → 9 ko 0 bana → first non-9 pe +1 → else new array"
```

## __Q7.Find Missing Number__

```python
#approach by the sum method__

def find(arr):
    n = len(arr)

    actual_sum = 0
    ##from 0 to n
    expected_sum = n*(n+1)//2

    for i in range(0,n):
        actual_sum = actual_sum + arr[i]

    return expected_sum - actual_sum

##driver code
arr = [0,1,2,4]
result = find(arr)
print('The missing number is:-',result)
```

```python
##another method of finding number without start from range 0 to n 
def find(arr):
    n = len(arr)

    expected_sum = (n+1)*(n+2)//2
    actual_sum = 0

    for i in range(n):
        actual_sum += arr[i]

    return expected_sum - actual_sum


# driver code
arr = [1,2,4]
print(find(arr))
```

## __Q8.IS_array Sorted?__

```python

def is_sorted(arr):
    n = len (arr)

    for i in range(0,n-1):
        if arr[i]>arr[i+1]:
            return False    
    else:
        return True
    

##driver code
arr = [5,4,3,2,1]
result = is_sorted(arr)
print(result)
```

## __Q.Move Zeroes__


```python
def move_zeroes(nums):
    j = 0  # position to place next non-zero
    n = len(nums)
    # Step 1: place all non-zero elements at front
    for i in range(n):
        if nums[i] != 0:
            nums[j], nums[i] = nums[i], nums[j]
            j += 1

    return nums


# driver code
nums = [0, 1, 0, 3, 12]
print(move_zeroes(nums))
```
