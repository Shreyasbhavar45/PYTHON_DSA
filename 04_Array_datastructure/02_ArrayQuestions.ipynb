## **Prerequisites to solve this question**

```python
# ================================
# ARRAY PREREQUISITES (BASICS)
# ================================

# 1. ARRAY TRAVERSAL
def traverse_array(arr):
    print("Traversal:")
    for i in range(len(arr)):
        print(f"Index {i} -> {arr[i]}")
    print()


# 2. FIND MAX ELEMENT
def find_max(arr):
    max_val = arr[0]
    for num in arr:
        if num > max_val:
            max_val = num
    return max_val


# 3. SUM OF ARRAY
def array_sum(arr):
    total = 0
    for num in arr:
        total += num
    return total


# 4. REVERSE ARRAY (TWO POINTER)
def reverse_array(arr):
    i = 0
    j = len(arr) - 1

    while i < j:
        arr[i], arr[j] = arr[j], arr[i]  # swap
        i += 1
        j -= 1

    return arr


# 5. MOVE ZEROES (TWO POINTER)
def move_zeroes(arr):
    j = 0  # position for non-zero

    for i in range(len(arr)):
        if arr[i] != 0:
            arr[j], arr[i] = arr[i], arr[j]
            j += 1

    return arr


# 6. MISSING NUMBER (SUM FORMULA)
def missing_number(arr):
    n = len(arr)
    expected = n * (n + 1) // 2
    actual = sum(arr)
    return expected - actual


# 7. PALINDROME CHECK (TWO POINTER)
def is_palindrome(arr):
    i = 0
    j = len(arr) - 1

    while i < j:
        if arr[i] != arr[j]:
            return False
        i += 1
        j -= 1

    return True


# ================================
# DRIVER CODE (TEST EVERYTHING)
# ================================
if __name__ == "__main__":
    arr = [1, 2, 0, 4, 0, 5]

    traverse_array(arr)

    print("Max Element:", find_max(arr))
    print("Sum:", array_sum(arr))
    print("Reverse:", reverse_array(arr.copy()))
    print("Move Zeroes:", move_zeroes(arr.copy()))
    print("Missing Number:", missing_number([0,1,3,4]))
    print("Is Palindrome:", is_palindrome([1,2,2,1]))
```

## __Q1.Maximum Element in List__



```python
##Approach 1 :-Brute Force Approach. Compare 1 ,1 element with each other.
##Functio Calling
def MaxElement(list1):
    n = len(list1)

    for i in range(n):
        ismax = True

        for j in range(n):
            if i == j:
                continue

            if list1[j]>list1[i]:
                ismax = False
                break

        if ismax:
            return list1[i]
        

##Driver code
list1 = [3,5,2,9,20,6]
#function Calling
Result = MaxElement(list1)
print("Maximum Element In the list is:-",Result)
```

```python
##Approach 2:- Optimal Approach. Assume the max element
#function Definition
def MaxElement(arr):
    n = len(arr)
    max = arr[0]
    for i in range(n):
        if arr[i]>max:
            max = arr[i]

    return max


##driver code
arr = [3,5,2,9,20,6]
#function Calling
Result = MaxElement(arr)
print("Maximum Element In the list is:-",Result)
```
