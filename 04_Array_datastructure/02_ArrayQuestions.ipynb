````python
##Approach 1 :-Brute Force Approach. Compare 1 ,1 element with each other.
##Function Calling
def MaxElement(list1):
    n = len(list1)#6

    for i in range(n):#0 to 5
        ismax = True

        for j in range(n):#0 to 5
            #it means when i = 0 and j = 0 the also continue the statement.
            # i = 1 and j = 1
            # i = 2 And j = 2
            # i = 3 and j = 3
            # i = 4 and j = 4
            # i = 5 and j = 5
            if i == j:
                continue

            if list1[j]>list1[i]:
                ismax = False
                break

        if ismax:
            return list1[i]
        

##Driver code
list1 = [3,5,2,9,20,6]
#function Calling
Result = MaxElement(list1)
print("Maximum Element In the list is:-",Result)
````

##
- left = 0 (start), right = last index

- Swap start and end elements
- Move pointers towards center

- Repeat until left < right

- Pointers cross → stop

````python
# 🔹 Function: Reverse a list in-place using Two Pointer Approach
#Time Complexity = O(n) Each element visited once.
#Space Complexity = O(1) No extra space (in-place).
def reverse_list(arr):

    # Step 1: Initialize two pointers
    left = 0                    # starting index
    right = len(arr) - 1        # last index

    # Step 2: Run loop until left < right
    # Reason: jab pointers cross ho jayenge tab swapping complete ho chuki hogi
    while left < right:

        # Step 3: Swap elements
        # Example: arr[left] ↔ arr[right]
        # [6,4,8,2,9,0] → swap 6 and 0 → [0,4,8,2,9,6]
        arr[left], arr[right] = arr[right], arr[left]

        # Step 4:  Two Move pointers # pointer approach
        left = left + 1         # move forward
        right = right - 1       # move backward

    # Step 5: Return reversed list
    return arr


# 🔹 Driver Code
arr = [6, 4, 8, 2, 9, 0]

# Function Calling
result = reverse_list(arr)

# Output
print("The reverse list is :-", result)
````

````python
# 🔹 Helper function: reverse part of array
def reverse(arr, start, end):
    while start < end:
        arr[start], arr[end] = arr[end], arr[start]
        start += 1
        end -= 1


# 🔹 Main function: rotate left by D positions
def rotate_left(arr, d):
    n = len(arr)

    # Edge case: if d > n
    ##“d represents the number of rotations, and we take d modulo n because rotating n times brings the array back to its original state.”
    d = d % n#16%5= 1 then the output of arr should be [2,3,4,5,1]

    # Step 1: Reverse first d elements
    reverse(arr, 0, d - 1)#index 0 or 1

    # Step 2: Reverse remaining elements
    reverse(arr, d, n - 1)#index [2 ,3, 4]

    # Step 3: Reverse whole array
    reverse(arr, 0, n - 1)

    return arr


# 🔹 Driver Code
arr = [1, 2, 3, 4, 5]
d = 16

result = rotate_left(arr, d)
print("Rotated Array:", result)
````

````python
class Solution(object):
    def rotate(self, nums, k):
        n = len(nums)
        k = k % n

        def reverse(left, right):
            while left < right:
                nums[left], nums[right] = nums[right], nums[left]
                left += 1
                right -= 1

        reverse(0, n - 1)
        reverse(0, k - 1)
        reverse(k, n - 1)


# driver code
nums = [1,2,3,4,5]
k = 2

obj = Solution()
obj.rotate(nums, k)

print(nums)
````

````python
def rotate_left(nums, k):
    n = len(nums)

    k = k % n   # handle large k

    # helper function
    def reverse(left, right):
        while left < right:
            nums[left], nums[right] = nums[right], nums[left]
            left += 1
            right -= 1

    # Step 1: reverse first k elements
    reverse(0, k - 1)

    # Step 2: reverse remaining elements
    reverse(k, n - 1)

    # Step 3: reverse whole array
    reverse(0, n - 1)

    return nums


# 🔹 input
nums = [1,2,3,4,5]
k = 2

# 🔹 function call
result = rotate_left(nums, k)

# 🔹 output
print("After left rotation:", result)
````

## __Q6.plus one.__


````python
#Approach
#method definition
def plus_one(arr):
    n = len(arr)
    right = len(arr)-1
    for i in range(right,-1,-1):
        if arr[i]==9:
            arr[i]=0

            continue

        elif arr[i] < 9:
            arr[i]+=1
            return arr
        
    return [1] + [0] * n

##driver code
arr = [1,2,3,4]
#function calling
result = plus_one(arr)
print(result)
````

````python
## good versionnn
# 🔹 Problem:
# Given an array of digits representing a number,
# add +1 to the number and return updated array.

# 🔹 Example:
# [1,2,3] → [1,2,4]
# [1,2,9] → [1,3,0]
# [9,9,9] → [1,0,0,0]


# 🔹 Approach (Important 🔥)

# Step 1:
# Right (last index) se start karna hai

# Step 2:
# Har digit check karna hai

# Step 3:
# Agar digit < 9 hai:
# → usme +1 karna hai
# → aur wahi stop (return array)

# Step 4:
# Agar digit == 9 hai:
# → usko 0 bana dena hai
# → carry next (left) element me jayega

# Step 5:
# Loop continue karega jab tak:
# → digit < 9 na mil jaye

# Step 6 (Special Case):
# Agar loop khatam ho gaya:
# → iska matlab sab digits 9 the
# → new array banega: [1] + [0]*n


# 🔹 Code (Final Correct Version)

def plus_one(arr):
    n = len(arr)

    for i in range(n-1, -1, -1):

        # Case 1: 9 hai → carry
        if arr[i] == 9:
            arr[i] = 0

        # Case 2: < 9 → +1 and stop
        else:
            arr[i] += 1
            return arr  # 🔥 yahi game changer hai turant arr ko stop krta hai

    # Case 3: All 9 (loop finish ho gaya)
    return [1] + [0] * n


# 🔹 Driver Code
arr = [1,2,3,4]
result = plus_one(arr)
print(result)


# 🔹 Complexity

# Time Complexity  = O(n)
# Space Complexity = O(1)


# 🔹 One-line Logic

# "Right se start → 9 ko 0 bana → first non-9 pe +1 → else new array"
````

## __Q7.Find Missing Number__

````python
#approach by the sum method__

def find(arr):
    n = len(arr)

    actual_sum = 0
    ##from 0 to n
    expected_sum = n*(n+1)//2

    for i in range(0,n):
        actual_sum = actual_sum + arr[i]

    return expected_sum - actual_sum

##driver code
arr = [0,1,2,4]
result = find(arr)
print('The missing number is:-',result)
````

````python
##without swap
class Solution(object):
    def moveZeroes(self, nums):

        j = 0

        # Step 1: copy non-zero elements
        for i in range(len(nums)):
            if nums[i] != 0:
                nums[j] = nums[i]
                j += 1

        # Step 2: fill remaining with zeros
        for i in range(j, len(nums)):
            nums[i] = 0

        return nums


# driver code
obj = Solution()
nums = [0,1,0,3,12]
result = obj.moveZeroes(nums)
print(result)
````

## __Majority Element__.

````python
def majority_element(nums):
    # Step 1: Boyer-Moore (find candidate)
    candidate = nums[0]
    count = 1

    for i in range(1, len(nums)):
        if nums[i] == candidate:
            count += 1
        else:
            count -= 1

        if count == 0:
            candidate = nums[i]
            count = 1

    # Step 2: Verification
    count = 0
    for num in nums:
        if num == candidate:
            count += 1

    if count > len(nums) // 2:
        return candidate
    else:
        return -1   # no majority element


# 🔹 driver code
nums = [4,4,1,4,2,3]
print(majority_element(nums))
````

````python
def majority_element(nums):
    """
    🔥 Boyer-Moore Voting Algorithm

    Goal:
    Find element jo n/2 se zyada baar aata hai (majority element)

    Core Idea:
    - Same element → vote increase (count++)
    - Different element → vote cancel (count--)
    - Jab count = 0 → new candidate choose karo

    """

    # Step 1: Start with first element as candidate
    candidate = nums[0]
    count = 1

    # Step 2: Traverse array
    for i in range(1, len(nums)):

        # Case 1: same element → support milta hai
        if nums[i] == candidate:
            count += 1

        # Case 2: different element → vote cancel hota hai
        else:
            count -= 1

        # Case 3: agar count 0 ho gaya
        # iska matlab previous candidate ka influence khatam
        if count == 0:
            candidate = nums[i]   # new candidate
            count = 1             # usko ek vote mila

    # Step 3: candidate hi answer hai (if majority exists)
    return candidate
    


# driver code
arr = [6,5,5]
result = majority_element(arr)
print(result)
````

## __Remove Duplicates in Sorted Array__

````python
# 🧠 Remove Duplicates (In-Place) — VS Code Notes

def removeDuplicates(nums):


# Step 0: Edge Case
# Agar array empty hai → koi element nahi → return 0
if len(nums) == 0:
    return 0


# Step 1: Initialize pointer2
# pointer2 → last unique element ka index
# Starting me first element always unique hota hai
pointer2 = 0


# Step 2: Traverse array
# i → scanning pointer (har element check karega)
# Loop 1 se start hoga kyunki 0 already handle ho chuka hai
for i in range(1, len(nums)):

    # Step 3: Compare current element with last unique element
    # nums[i] → current element
    # nums[pointer2] → last unique element

    # Case 1: Duplicate
    # Agar dono same hain → skip (kuch nahi karna)
    # pointer2 move nahi karega

    # Case 2: New Element
    # Agar different hai → naya element mila

    if nums[i] != nums[pointer2]:

        # Step 4: Move pointer2 forward
        # Next position ready karo
        pointer2 += 1

        # Step 5: Place new element
        # Current element ko next unique position pe likh do
        nums[pointer2] = nums[i]


# Step 6: Return result
# pointer2 index hai → count ke liye +1
return pointer2 + 1
```

# 🔍 Dry Run Example

# Input:

# nums = [1,1,2,2,3]

# Start:

# pointer2 = 0 → [1,*,*,*,*]

# i = 1 → 1 == 1 → duplicate → skip

# i = 2 → 2 != 1 → pointer2 = 1 → nums[1] = 2 → [1,2,*,*,_]

# i = 3 → 2 == 2 → duplicate → skip

# i = 4 → 3 != 2 → pointer2 = 2 → nums[2] = 3 → [1,2,3,*,*]

# Final:

# k = 3

# nums = [1,2,3,2,3]  (first k elements important)

# 🎯 Key Points

# pointer2 → last unique element track karta hai

# i → array scan karta hai

# duplicate → skip

# new element → aage likho

# in-place → same array modify hota hai

# 🧠 One Line Summary

# "Scan karo, compare karo, new ho toh aage likho"
````

````python
def removeDuplicates(nums):
    if len(nums) == 0:
        return 0

    pointer2 = 0

    for i in range(1, len(nums)):
        if nums[i] != nums[pointer2]:
            pointer2 += 1
            nums[pointer2] = nums[i]

    return pointer2 + 1


# driver code
nums = [1,1,2,2,3]
k = removeDuplicates(nums)

print(k)        # 3
print(nums)     # [1,2,3,2,3] (first k elements important)
````

## __Best time to buy and sell stock__

````python
##Best time to buy and sell the stocks.
##time complexity :- O(n)
##function Definition 
def FindMaxProfit(price):
    #initialization.
    minprice = float('inf')#infinite value,Taaki first element automatically minimum ban jaye.
    maxprofit = 0

    #Ye loop har price ko ek baar check karta hai.
    for i in range(len(price)):

                # -------- Day 1 --------
        # price = 7
        # 7 < inf → True
        # isliye minprice = 7 ho jayega
        # profit calculate nahi hoga kyuki abhi buy price hi set hua hai

        # -------- Day 2 --------
        # price = 4
        # 4 < 7 → True
        # minprice update hoga
        # minprice = 4

        # -------- Day 3 --------
        # price = 5
        # 5 < 4 → False
        # profit check karenge
        # profit = 5 - 4 = 1
        # maxprofit = 1

        # -------- Day 4 --------
        # price = 3
        # 3 < 4 → True
        # minprice update
        # minprice = 3

        # -------- Day 5 --------
        # price = 6
        # 6 < 3 → False
        # profit = 6 - 3 = 3
        # maxprofit = 3

        # -------- Day 6 --------
        # price = 16
        # 16 < 3 → False
        # profit = 16 - 3 = 13
        # maxprofit = 13

        if price[i]< minprice:
            minprice = price[i]
        
        elif price[i] - minprice > maxprofit:
            maxprofit = price[i]-minprice
        
    return maxprofit



##driver code
price = [7,4,5,3,6,16]
maxprofit_value = FindMaxProfit(price)
print('the max profit of buying and selling stock is:-',maxprofit_value)

'''We iterate through the array once and keep track of the minimum price seen so far.
For each day we calculate the profit if we sell today.
We update the maximum profit accordingly.
Since we traverse the array only once, the time complexity is O(n) and space complexity is O(1).'''
````

## __Product of array except self__

````python
##brute force approach
##method definition
##time complexity = O(n^2)
##space complexity = O(n)
def productExceptSelf(nums):

    #new array define for store the new values.
    #initial
    new_arr = []
    
    #outerloop
    for i in range(len(nums)):
        product = 1

        #inner loop
        for j in range(len(nums)):

            #condition
            #if 0 == 0 skip
            #if 1 == 1 skip .. like so we can skip the product
            
            if i == j :
                continue
            product = product * nums[j]


        new_arr.append(product)

    return new_arr

##driver code
nums = [1,2,3,4,5]
#function calling
result = productExceptSelf(nums)
print(result)
````

````python
class Solution:
    def productExceptSelf(self, nums):
        n = len(nums)

        # Step 1: answer array (left products)
        answer = [1] * n

        # Step 2: build left products
        for i in range(1, n):
            answer[i] = answer[i-1] * nums[i-1]

        # Step 3: right product variable
        right = 1

        # Step 4: traverse from right
        for i in range(n-1, -1, -1):
            answer[i] = answer[i] * right
            right = right * nums[i]

        return answer
    
##driver code
nums = [1,2,3,4,5]
#function callingg
result = productExceptSelf(nums)
print(result)
````
